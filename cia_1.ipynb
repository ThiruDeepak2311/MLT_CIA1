```python
import random as rd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#1 Generating 1000 random numbers

np.random.seed(23)
x = np.random.randn(1000)
y = np.random.randn(1000) 
```

```python
plt.scatter(x,y)
```

```python
#2 Simple Linear Regression

x1=x-x.mean()
y1=y-y.mean()
num=x1*y1 
den=x1*x1
m=num.sum()/den.sum()
```

```python
print("Slope is : ",m)
```

```python
c=y.mean()-(m*x.mean())
print('C Value: ',c)
```

```python
ycap=m*x+c
e=y-ycap
mse=(e*e).sum()
print('MeanSquaredError:',mse)
```

```python
#3 Gradient Descent

p = 0
q = 0

learnig_rate = 0.01 #LEARNING RATE

iterations = 1000 #number of iterations

for i in range(iterations):
    y_pred = q * x + p
    err = y - y_pred
    p_grad = -(1/len(x)) * sum(err)
    q_grad = -(1/len(x)) * sum(err * x)
    p = p - learnig_rate * p_grad
    q = q - learnig_rate * q_grad

print("Intercept p:", p)#INTERCEPT
print("Slope q:", q)#SLOPE
```

```python
cost = (1/(2*len(x))) * sum(err**2)
print('Mean Squared error is: ',cost)
```

```python
#4 Best Fit Line

y1  = q*x+p 
plt.scatter(x,y)
plt.plot(x,y1,"red")
```

```python
#5 Using sklearn

x=x.reshape(-1,1)
y=y.reshape(-1,1)
from sklearn.linear_model import LinearRegression
reg=LinearRegression()
reg.fit(x,y)
```

```python
print('Intercept using sklearn: ',reg.intercept_)
print('Slope using sklearn: ' ,reg.coef_)
print('Intercept using Gradient Descent  :',p.round(8))
print('Slope using Gradient Descent      :',q.round(8))
```
